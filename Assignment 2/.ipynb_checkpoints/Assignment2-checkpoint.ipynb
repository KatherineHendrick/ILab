```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
% matplotlib inline
from matplotlib import rc
rc('text', usetex=True)
from pylab import rcParams
rcParams['figure.figsize'] = 10,8

#imports all needed packages
```

```python
def linearFit(t, m, b): 
    return m*t + b

def quadraticFit(t, a, b, c):
    return a*(t**2)+(b*t)+c

def expDecay(t, a, b, c):
    return a*np.exp(-b*t) + c

def gaussianWithOffset(t, a, c, s):
    return a*np.exp(-((t-c)**2)/(2*s**2))

def cosine (t, a, w, c):
    return a*(np.cos(w*(t-c)))

#defines needed functions
```

```python
data = np.genfromtxt('002data.txt', skip_header=1)
#retrieves data from saved file
```

```python
data[ : ,0]  
```

```python
x = data[ : ,0]
z = data[ : ,1]
u = data[ : ,2]
v = data[ : ,3]
y = data[ : ,4]
eVar = data[ : ,5]

#references columns from the data table for easier use
```

## Question 1

```python
fitParams, fitCovariances = curve_fit(linearFit, x, z)
print(fitParams)
print(fitCovariances)
#sets the fit parameters to follow the previously defined linear fit
```

```python
plt.plot(x,z,'o')
plt.title('Distance as a Function of Time', fontsize = 20)
plt.ylabel(r'$\mathrm{Z (m)}$', fontsize = 18)
plt.xlabel(r'$\mathrm{x (s)}$', fontsize = 18)

plt.plot(x, linearFit(x, fitParams[0], fitParams[1]))
plt.show()
```

The above plot shows a straight line (of the form $y=mx+b$) best fit of the z vs x data from the given data set.

# Question 2

```python
plt.loglog(x,u,'o')
#plots data with both axes logarithmic
plt.title('Distance vs Time (Logarithmic Axes))', fontsize = 20)
plt.ylabel(r'$\mathrm{U (m)}$', fontsize = 18)
plt.xlabel(r'$\mathrm{X (s)}$', fontsize = 18)
plt.show()
```

The above plot of the u vs x data shows a linear fit when plotted with both the x and y axes logarithmic. This indicates that the data set fits the form $U=ax^k$, where a and k are constants.

```python
plt.semilogy(x,v,'o')
#plots data with a logarithmic y-azis
plt.title('Distance vs Time (Logarithmic y-axis))', fontsize = 20)
plt.ylabel(r'$\mathrm{V (m)}$', fontsize = 18)
plt.xlabel(r'$\mathrm{X (s)}$', fontsize = 18)
plt.show()
```

Contrary to the previous example, the above graph shows the v vs. x data graphed with only the v axis plotted logarithmically.  This indicates that x is in the exponent of the line's equation (eg $V = ka^{mx}$, with k, a, and m as constants.) 

# Question 3

```python
plt.plot(x,y,'o')
plt.title('Plot of Distance vs Time (With Error Bars)', fontsize = 20)
plt.ylabel(r'$\mathrm{y (m) \pm e(m)}$', fontsize = 18)
plt.xlabel(r'$\mathrm{x (s)}$', fontsize = 18)

plt.errorbar(x, y, fmt = 'ro', yerr = eVar)
# plots vertical error bars of +/- the e column in the data set

plt.show()
```

Above, I plotted y vs x with error bars $\pm e_i$, the corresponding uncertainty for each data point.

# Question 4

```python
fitParams2, fitCovariances2 = curve_fit(quadraticFit, x, y)
print(fitParams2)
print(fitCovariances2)
#sets the fit parameters to follow the previously defined quadratic fit
```

```python
plt.plot(x,y,'o')
plt.title('Plot of Distance vs Time (quadratic fit)', fontsize = 20)
plt.ylabel(r'$\mathrm{y (m) \pm e(m)}$', fontsize = 18)
plt.xlabel(r'$\mathrm{x (s)}$', fontsize = 18)

plt.errorbar(x, y, fmt = 'ro', yerr = eVar)
#plots error bars

sigmaQ = np.sqrt([fitCovariances2[0,0], fitCovariances2[1,1], fitCovariances2[2,2]])
#defines sigma

plt.plot(x, quadraticFit(x, fitParams2[0], fitParams2[1], fitParams2[2]), \
         label='Best Fit: $y = %.2f x^2 + %.2f x + %.2f$' %(fitParams2[0], fitParams2[1], fitParams2[2]))
plt.plot(x, quadraticFit(x, fitParams2[0] + 3*sigmaQ[0], fitParams2[1] - 3*sigmaQ[1] , fitParams2[2] - 3*sigmaQ[2]), \
         label='Low Fit: $y = (%.2f + 3 \sigma_a) x^2 + (%.2f -3 \sigma_b) x + (%.2f-3 \sigma_c)$' \
         %(fitParams2[0]+3*sigmaQ[0], fitParams2[1]- 3*sigmaQ[1] , fitParams2[2] - 3*sigmaQ[2]))
plt.plot(x, quadraticFit(x, fitParams2[0] - 3*sigmaQ[0], fitParams2[1] + 3*sigmaQ[1], fitParams2[2] + 3*sigmaQ[2]), \
         label='High Fit: $y = (%.2f - 3 \sigma_a) x^2 + (%.2f +3 \sigma_b) x + (%.2f+3 \sigma_c)$' \
         %(fitParams2[0]-3*sigmaQ[0], fitParams2[1]+ 3*sigmaQ[1] , fitParams2[2] + 3*sigmaQ[2]))
plt.legend(bbox_to_anchor=(0.9,0.2), fontsize=15)
plt.show()

```

Above, you can see the plot from question 3, with a quadratic best fit line of the form $y=ax^2+bx+c$, and the corresponding high and low lines of best fit.  The high and low fit lines are within 3 standard deviations, determined from the standard deviations of each constant, a, b, and c.

# Question 5

```python
fitParams3, fitCovariances3 = curve_fit(gaussianWithOffset, x, y)
print(fitParams3)
print(fitCovariances3)

#sets the fit parameters to follow the previously defined gaussian fit
```

```python
plt.plot(x,y,'o')
plt.title('Plot of Distance vs Time (gaussian fit)', fontsize = 20)
plt.ylabel(r'$\mathrm{y (m) \pm e(m)}$', fontsize = 18)
plt.xlabel(r'$\mathrm{x (s)}$', fontsize = 18)
plt.errorbar(x, y, fmt = 'ro', yerr = eVar)
sigmaG = np.sqrt([fitCovariances3[0,0], fitCovariances3[1,1], fitCovariances3[2,2]])

plt.plot(x, gaussianWithOffset(x, fitParams3[0], fitParams3[1], fitParams3[2]),\
         x, gaussianWithOffset(x, fitParams3[0] + 3*sigmaG[0], fitParams3[1] + 3*sigmaG[1], fitParams3[2] + 3*sigmaG[2]),\
         x, gaussianWithOffset(x, fitParams3[0] + 3*sigmaG[0], fitParams3[1] + 3*sigmaG[1], fitParams3[2] - 3*sigmaG[2]),\
         x, gaussianWithOffset(x, fitParams3[0] + 3*sigmaG[0], fitParams3[1] - 3*sigmaG[1], fitParams3[2] + 3*sigmaG[2]),\
         x, gaussianWithOffset(x, fitParams3[0] + 3*sigmaG[0], fitParams3[1] - 3*sigmaG[1], fitParams3[2] - 3*sigmaG[2]),\
         x, gaussianWithOffset(x, fitParams3[0] - 3*sigmaG[0], fitParams3[1] + 3*sigmaG[1], fitParams3[2] + 3*sigmaG[2]),\
         x, gaussianWithOffset(x, fitParams3[0] - 3*sigmaG[0], fitParams3[1] - 3*sigmaG[1], fitParams3[2] + 3*sigmaG[2]),\
         x, gaussianWithOffset(x, fitParams3[0] - 3*sigmaG[0], fitParams3[1] - 3*sigmaG[1], fitParams3[2] - 3*sigmaG[2]),\
         )

plt.show()
```

The above plot fits the same data to a gaussian curve, of the form $y = ae^{-\frac{x-c}{2\mu^2.  The high and low lines of best fit are more difficult to determine here, as there isn't a clear 'high' or 'low' value from all of the possible lines of fit. 

```python
fitParams4, fitCovariances4 = curve_fit(cosine, x, y)
print(fitParams4)
print(fitCovariances4)
```

```python
plt.plot(x,y,'o')
plt.title('Plot of Distance vs Time (cosine fit)', fontsize = 20)
plt.ylabel(r'$\mathrm{y (m) \pm e(m)}$', fontsize = 18)
plt.xlabel(r'$\mathrm{x (s)}$', fontsize = 18)
#plt.xlim(0,3.5)
# plot the data as red circles with errorbars in the vertical direction
plt.errorbar(x, y, fmt = 'ro', yerr = eVar)
# now plot the best fit curve and also +- 3 sigma curves
# the square root of the diagonal covariance matrix element 
# is the uncertianty on the corresponding fit parameter.
sigmaC = np.sqrt([fitCovariances4[0,0], fitCovariances4[1,1], fitCovariances4[2,2]])

plt.plot(x, cosine(x, fitParams4[0], fitParams4[1], fitParams4[2]),\
         x, cosine(x, fitParams4[0] + 3*sigmaC[0], fitParams4[1] + 3*sigmaC[1], fitParams4[2] + 3*sigmaC[2]),\
         x, cosine(x, fitParams4[0] + 3*sigmaC[0], fitParams4[1] + 3*sigmaC[1], fitParams4[2] - 3*sigmaC[2]),\
         x, cosine(x, fitParams4[0] + 3*sigmaC[0], fitParams4[1] - 3*sigmaC[1], fitParams4[2] + 3*sigmaC[2]),\
         x, cosine(x, fitParams4[0] + 3*sigmaC[0], fitParams4[1] - 3*sigmaC[1], fitParams4[2] - 3*sigmaC[2]),\
         x, cosine(x, fitParams4[0] - 3*sigmaC[0], fitParams4[1] + 3*sigmaC[1], fitParams4[2] + 3*sigmaC[2]),\
         x, cosine(x, fitParams4[0] - 3*sigmaC[0], fitParams4[1] - 3*sigmaC[1], fitParams4[2] + 3*sigmaC[2]),\
         x, cosine(x, fitParams4[0] - 3*sigmaC[0], fitParams4[1] - 3*sigmaC[1], fitParams4[2] - 3*sigmaC[2]),\
        )
plt.show()
```

The above plot fits the data once again, but to a cosine function , of the form $y = a cos(\omega(x-c))$, again with difficulty determining the high and low fit lines.
